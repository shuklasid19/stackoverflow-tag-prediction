```python
#it will divide the classes same for both
x_train ,x_test,  y_train , y_test = train_test_split(x , y, random_state=101)
```

```python
#build model
```

```python
#instantiating the classifiers
sgd = SGDClassifier()
lr = LogisticRegression(solver='lbfgs')
svc = LinearSVC()
```

```python
#jaccard metric
def j_score(y_true , y_pred):
  jaccard = np.minimum(y_true, y_pred).sum(axis=1)/np.maximum(y_true, y_pred).sum(axis=1)
  return jaccard.mean()*100

#return which classifier using here onevsrestclassifier
def print_score(y_pred, clf):
  print("clf", clf.__class__.__name__)
  print('jacard score: {}'.format(j_score(y_test, y_pred)))
  print('-----')
```

```python
#it will iterate over all the algo in our list
#we can not use normal metrics for multilabel classification
#we need 

for classifier in [sgd , lr, svc]:
  clf = OneVsRestClassifier(classifier)
  clf.fit(x_train ,y_train)
  y_pred = clf.predict(x_test)
  print_score(y_pred, classifier)
  
```

```python

for classifier in [LinearSVC(C=1.5)]:
  clf = OneVsRestClassifier(classifier)
  clf.fit(x_train ,y_train)
  y_pred = clf.predict(x_test)
  print_score(y_pred, classifier)
```

```python

for classifier in [LinearSVC(C=1.5, class_weight='balanced' , penalty='l2')]:
  clf = OneVsRestClassifier(classifier)
  clf.fit(x_train ,y_train)
  y_pred = clf.predict(x_test)
  print_score(y_pred, classifier)
```

```python

for classifier in [LinearSVC(C=1.5, class_weight='balanced', penalty='l1' , dual=False)]:
  clf = OneVsRestClassifier(classifier)
  clf.fit(x_train ,y_train)
  y_pred = clf.predict(x_test)
  print_score(y_pred, classifier)
```
